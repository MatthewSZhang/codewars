```r
# kata 1
your_order <- function(s){
  s_splitted <- strsplit(s, ' ')[[1]]
  num <- as.numeric(gsub("\\D", "", s_splitted))
  ind = order(num)
  r <- vector()
  for (p in ind){
      r <- c(r, s_splitted[p])
  }
  your_order <- paste(r,collapse=' ')
}

your_order <- function(s){
  l <- unlist(strsplit(s, " "))
  names(l) <- as.numeric(gsub("[[:alpha:]]", "", l)) # creat a Primitive function to call C code compiled into R at build time
  paste(l[order(names(l))], collapse = " ")
}
```

```r
# kata 2
race <- function (v1, v2, g) {
  if (v1 > v2) {
    return(NULL)
  }
  t = g/(v2-v1)
  hh = as.integer(t)
  if (t-hh > 0.99999) {
      hh = hh+1
  }
  t = (t - hh)*60
  mm = as.integer(t)
  if (t-mm > 0.99999) {
      mm = mm+1
  }
  t = (t - mm)*60
  ss = as.integer(t)
  if (t-ss > 0.99999) {
      ss = ss+1
  }
  c(hh, mm, ss)
}


race <- function (v1, v2, g) {
    if (v1 < v2){
        sec <- 3600*g/(v2 - v1)
        c(sec%/%3600, sec%%3600%/%60, as.integer(sec%%3600%%60))
    }
    else { NULL }
}
```

```r
# kata 3
high <- function(x){
  l = unlist(strsplit(x, " "))
  m = 0
  for (p in l) {
      temp = sum(match(unlist(strsplit(p, split="")), letters))
      if (temp > m) {
          m = temp
          r = p
      }
  }
  return(r)
}

high <- function(s){
  scores <- sapply(unlist(strsplit(casefold(s), " ")), function(x) sum(utf8ToInt(x) - 96))
  names(scores)[which.max(scores)] # named vector e.g. v = c("Mary", "Sue"); names(v) = c("First", "Last")
}
```

```r
# kata 4
inArray <- function(a1, a2) {
  r = vector()
  for (p in a1[order(a1)]) {
      if (sum(grepl(p, a2))) {
          r = c(r, p)
      }
  }
  if (length(r) == 0) {
    return(NULL)
  }
  return(unique(r))
}

inArray <- function(a1, a2) {
  res <- unique(sort(a1[vapply(a1, function(s) any(grepl(s, a2)), USE.NAMES = FALSE, FUN.VALUE = logical(1))]))
  if (length(res) == 0) NULL
  else res
}
```

```r
# kata 5
digital_root <- function(n){
    ch = unlist(strsplit(as.character(n), ""))
    num = as.numeric(ch)
    r = sum(num)
    if (r > 10) {
        digital_root(r)
    } else {
        return(r)
    }
}

digital_root <- function(n){
  while (n > 9)
    n <- sum(as.numeric(strsplit(as.character(n), "")[[1]]))
  n
}
```

```r
# kata 6
phone <- function(strings, num) {
  vstrings = unlist(strsplit(strings, '\n'))
  s = grep(num, vstrings, value = TRUE)
  if (length(s) > 1) {
      r = paste0("Error => Too many people: ", num)
  } else if (length(s) == 0) {
      r = paste0("Error => Not found: ", num)
  } else {
      name = regmatches(s, regexpr("(?<=\\<).*?(?=\\>)", s, perl=T))
      vstr = unlist(strsplit(s, " "))
      phone_ind = grep(num, vstr)
      addr = paste(vstr[-phone_ind], collapse = ' ')
      addr = sub(paste0('<', name, '>'), '', addr)
      addr = sub('_', ' ', addr)
      addr = gsub('\\s+', ' ', addr)
      addr = trimws(gsub("(?![-.])[[:punct:]]", "", addr, perl=TRUE))
      name = trimws(name)
      r = paste0('Phone => ', num, ', Name => ', name, ', Address => ', addr)
  }
}


phone <- function(strng, num) {
    lines <- strsplit(strng,"\n")[[1]]
    line <- grep(num,lines,value = TRUE)
    if(length(line)==0) return(paste("Error => Not found:",num))
    nb <- sum(grepl("<.*?>",line))
    if (nb>1) return(paste("Error => Too many people:", num))
    nm <- gsub(".*?<(.*?)>.*","\\1",line)
    address <- trimws(gsub("\\s+"," ",gsub("[/+;\\$\\*\\?,:]","",sub("_"," ",sub("<(.*)>","",sub(num,"",line))))))
    sprintf("Phone => %s, Name => %s, Address => %s",num,nm,address)
}

# * is greedy matching that tries to match maximal possible by use *? change to minimal possible. e.g.
line = 'asdf<dfda<5616>sda'
a = c(gsub(".*?<(.*)>.*","\\1",line), gsub(".*<(.*)>.*","\\1",line))
names(a) <- c("lazy", "greedy")
print(a)
```
