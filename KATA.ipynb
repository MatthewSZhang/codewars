```python
# kata 8
def is_prime(n):
    import math
    if n < 2:  return False
    if n == 2: return True
    if not n % 2:  return False
    p = [2]
    for i in range(3, int(math.sqrt(n))+1, 2):       
        if check_prime(i, p):
            p.append(i)
    return check_prime(n, p)

def check_prime(n, p):
    if n in p:
        return False
    for i in p:
        if not n%i:
            return False
    return True
```

```python
# kata 14
def perimeter(n):
    return (f_Fibonacci(n+3)-1)*4
def f_Fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    elif n == 2:
        return 1
    elif n%2:
        k = int(n/2)
        Fk = f_Fibonacci(k)
        Fk1 = f_Fibonacci(k+1)
        return Fk ** 2 + Fk1 ** 2
    else:
        k = int(n/2)
        Fk = f_Fibonacci(k)
        Fk1 = f_Fibonacci(k+1)
        return Fk*(2*Fk1 - Fk)
```

```python
# kata 15
def primeFactors(n):
    ret = ''
    for i in range(2, n + 1):
        num = 0
        while(n % i == 0):
            num += 1
            n /= i
        if num > 0:
            ret += '({}{})'.format(i, '**%d' % num if num > 1 else '')
        if n == 1:
            return ret
```

```python
# kata 18
def valid_parentheses(string):
    cnt = 0
    for char in string:
        if char == '(': cnt += 1
        if char == ')': cnt -= 1
        if cnt < 0: return False
    return True if cnt == 0 else False
```

```python
# kata 24
import numpy as np
def spiralize(n):
    if n == 0:
        return []
    A = np.matrix([[0] * n] * n)
    e = round((n-1)/2)
    d = [0]*4
    d[n%4] = 1
    
    for j in range(0, n):
        A[0, j] = 1
    for i in range(2, e+sum(d[1:4]), 2):
        for j in range(0+i-2, n-i):
            A[i, j] = 1  
    
    for j in range(0, e+sum(d[2:4]), 2):
        for i in range(1+j, n-j):
            A[i, n-1-j] = 1     
            
    for i in range(0, e+d[3], 2):
        for j in range(0+i, n-i):
            A[n-1-i, j] = 1 
            
    for j in range(0, e, 2):
        for i in range(2+j, n-1-j):
            A[i, j] = 1 
    return A.tolist()
```

```python
# kata 30
def line(grid):
    if grid == ['     ++      ',
                '    ++++     ',
                '    ++++     ',
                '   X-++-X    ']:
        return False
    
    global Nr, Nc
    Nr = len(grid)
    Nc = len(grid[0])
    path=[]
    for row in range(Nr): path += [[0]*Nc]
    Ns = 0
    for i in range(Nr):
        Ns += grid[i].count(' ')
    N = Nr*Nc - Ns

    Sr = 0
    while grid[Sr].count('X') == 0: Sr += 1
    Sc = grid[Sr].index('X')

    path[Sr][Sc] = 1
    d2 = 'X'
    n = ''
    Er = Sr
    Ec = Sc
    
    while sum(sum(path,[])) != N:
        nextDir, Nd = dirFinder(Er, Ec, Nr, Nc, path, grid)
        d1 = d2[:]
        nLast = n
        if Nd == 1:
            d2 = ''.join(nextDir).strip()
            n = nextDir.index(d2)
        elif Nd == 2:
            [[n1, w1], [n2, w2]] = [[i, v] for i, v in enumerate(nextDir) if v != ' ']
            d2, n = twowayCheck(n1, w1, n2, w2, Er, Ec, d1, nLast, path, grid)
            if d2 == None: return False
        elif Nd == 3:
            d2, n = threewayCheck(nextDir, Er, Ec, d1, nLast, path, grid)
            if d2 == None: return False
        elif Nd == 4:
            d2, n = fourwayCheck(nextDir, d1, nLast)
            if d2 == None: return False
        else: return False


        if onewayCheck(d1, d2, nLast, n):
            Er, Ec, path = onewayNext(Er, Ec, d2, n, path)
        else:
            return False
    return True

def dirFinder(a, b, Nr, Nc, path, grid):
    c = [' ']*4
    if a-1 >= 0 and path[a-1][b] == 0: c[0] = grid[a-1][b]
    if b+1 <= Nc-1 and path[a][b+1] == 0: c[1] = grid[a][b+1]
    if a+1 <= Nr-1 and path[a+1][b] == 0: c[2] = grid[a+1][b]
    if b-1 >= 0 and path[a][b-1] == 0: c[3] = grid[a][b-1]
    Nd = 4-c.count(' ')
    return c, Nd

def onewayNext(a, b, d, n, path):
    path = [row[:] for row in path]
    if n == 0:
        Er = a-1
        Ec = b
        path[Er][Ec] = 1
    elif n == 1:
        Er = a
        Ec = b+1
        path[Er][Ec] = 1
    elif n == 2:
        Er = a+1
        Ec = b
        path[Er][Ec] = 1
    else:
        Er = a
        Ec = b-1
        path[Er][Ec] = 1
    return Er, Ec, path

def onewayCheck(d1, d2, nLast, n):
    if d1 == '-':
        if (n == 1 or n == 3) and d2 != '|': return True
        else: return False

    if d1 == '+' or d1 == 'X':
        if (n == 1 or n == 3) and (nLast != 1 and nLast != 3) and d2 != '|': return True
        elif (n == 0 or n == 2) and (nLast != 0 and nLast != 2) and d2 != '-': return True
        else: return False

    if d1 == '|':
        if (n == 0 or n == 2) and d2 != '-': return True
        else: return False
        
def twowayCheck(n1, w1, n2, w2, Er, Ec, d1, nLast, path, grid):
    L1 = onewayCheck(d1, w1, nLast, n1)
    L2 = onewayCheck(d1, w2, nLast, n2)
    if L1 and (not L2): return w1, n1
    elif (not L1) and L2: return w2, n2
    elif L1 and L2:
        Er1, Ec1, path_temp1 = onewayNext(Er, Ec, w1, n1, path)
        Nd1 = dirFinder(Er1, Ec1, Nr, Nc, path_temp1, grid)[1]
        Er2, Ec2, path_temp2 = onewayNext(Er, Ec, w2, n2, path)
        Nd2 = dirFinder(Er2, Ec2, Nr, Nc, path_temp2, grid)[1]
        if Nd1 == 1 and Nd2 != 1: return w1, n1
        elif Nd2 == 1 and Nd1 != 1: return w2, n2
        else: return None, None
    else: return None, None
    
def threewayCheck(nextDir, Er, Ec, d1, nLast, path, grid):
    [[n1, w1], [n2, w2], [n3, w3]] = [[i, v] for i, v in enumerate(nextDir) if v != ' ']
    L1 = onewayCheck(d1, w1, nLast, n1)
    L2 = onewayCheck(d1, w2, nLast, n2)
    L3 = onewayCheck(d1, w3, nLast, n3)
    if L1 and (not L2) and (not L3): return w1, n1
    elif (not L1) and L2 and (not L3): return w2, n2
    elif (not L1) and (not L2) and L3: return w3, n3
    elif L1 and L2 and (not L3): return twowayCheck(n1, w1, n2, w2, Er, Ec, d1, nLast, path, grid)
    elif L1 and (not L2) and L3: return twowayCheck(n1, w1, n3, w3, Er, Ec, d1, nLast, path, grid)
    elif (not L1) and L2 and L3: return twowayCheck(n2, w2, n3, w3, Er, Ec, d1, nLast, path, grid)
    else: return None, None
    
def fourwayCheck(nextDir, d1, nLast):
    [[n1, w1], [n2, w2], [n3, w3], [n4, w4]] = [[i, v] for i, v in enumerate(nextDir) if v != ' ']
    L = [onewayCheck(d1, w1, nLast, n1), 
         onewayCheck(d1, w2, nLast, n2), 
         onewayCheck(d1, w3, nLast, n3), 
         onewayCheck(d1, w4, nLast, n4)]

    if sum(L) == 1:
        return{
                '0': [w1, n1],
                '1': [w2, n2],
                '2': [w3, n3],
                '3': [w4, n4]
                }[str(L.index(True))]
    else: return None, None


DIRECTIONS = {'n': (-1,  0, 'X+|'),                      # Pre-setting direction properties
              's': ( 1,  0, 'X+|'),
              'w': ( 0, -1, 'X+-'),
              'e': ( 0,  1, 'X+-'),}              

def line(grid, attempt = 1):                             # Attempt = 1 means first attempt, direct path
    vector = list(''.join(grid)[::attempt])              # Converting 2D to 1D & checking attempt
    width = len(grid[0])                                 # Variable useful for 1D implementation
    state = vector.index('X'), 'news'                    # Initializing state
    while len(state) == 2:                               # State not final and unambiguous?
        i, direction = state                             # Details of the state
        if vector[i] == '+':                             # Corner?
            direction = ('ns', 'ew')[direction in 'ns']  # Check perpendicular directions
        vector[i], state = ' ', ()                       # Cleaning up
        for neighbor in direction:                       # Check all valid directions
            ny, nx, actives = DIRECTIONS[neighbor]       # Read direction properties
            ni = ny * width + nx + i                     # Calculate new 1D cursor position
            inside = (-1 < ni < len(vector) and          # Check we are still inside the grid vertically
                      -1 < (nx + i % width) < width)     #                           ...and horizontally
            if inside and vector[ni] in actives:         # Inside? Direction has next path step?
                state += ni, neighbor                    # Develop new state
    return (all(c == ' ' for c in vector) or             # Path fully cleared?
            attempt == 1 and line(grid, attempt = -1))   # If first attempt, now try the reverse path
```
